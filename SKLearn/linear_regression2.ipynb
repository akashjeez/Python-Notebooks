```python
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
import seaborn as sn
```

```python
## Linear Regression - Multiclass Classification
digits = load_digits()
dir(digits)
```

```python
digits.data[0]
```

```python
plt.gray()
for i in range(10):
    plt.matshow(digits.images[i])
```

```python
digits.target
```

```python
digits.target_names
```

```python
# train dataset is 80% and test dataset is 20%, both will choose random data.
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2)

print(f"Total Count of Dataset : {len(digits.data)}")
print(f"Count of Training Dataset : {len(X_train)}")
print(f"Count of Testing Dataset : {len(X_test)}")
```

```python
## Linear Regression Model - Single Variable
model = LogisticRegression()
## fit() - training linear regression model using available data points.
model.fit(X_train, y_train)
```

```python
# Calculate Accuracy of the trained dataset.
accuracy = model.score(X_test, y_test)
print(f"Accurancy: {round(accuracy * 100, 2)} %")
```

```python
digits.target[67]
```

```python
model.predict( [digits.data[67]] )
```

```python
model.predict( digits.data[0:5] )
```

```python
y_predicted = model.predict(X_test)
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
plt.figure(figsize = (10, 7))
sn.heatmap(cm, annot = True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
